## Ejercicio 2
Escribe un programa para pasar de grados a radianes. Hay que usar `input`. Recuerda que la conversión se realiza mediante

radianes = grados*(pi/180)

```python
print("Este programa convierte grados a radianes")
print("Introduzca los grados a convertir")
grados = float(input())

pi = 3.14159265359

radianes = grados*(pi/180)

print(f"{grados} grados = {radianes} radianes")
```

## Ejercicio 3
Escribe un programa que calcule el area de un paralelogramo (base x altura). También con `input`

```python
print("Este programa calcula el area de un paralelograma")
print("Indique la base y altura del paralelograma")
base = float(input("Base: "))
altura = float(input("Altura: "))

area = base * altura

print(f"El area del paralelograma es {area}")
```

## Ejercicio 4
Escribe un programa que calcule el area de un paralelogramo (base x altura). También con `input`. En este caso debe ser capaz de admitir valores reales, con decimales, para base y altura en español. (PISTA: Recuerda que tendrás que hacer uso del método replace)

```python
print("Este programa calcula el area de un paralelograma")
print("Indique la base y altura del paralelograma")
base = input("Base: ")
altura = input("Altura: ")

nueva_base = float(base.replace(",", "."))
nueva_altura = float(altura.replace(",", "."))

area = nueva_base * nueva_altura

print(f"El area del paralelograma es {area}")
```
