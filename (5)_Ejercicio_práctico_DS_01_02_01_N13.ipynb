* Aplícalo para encontrar los duplicados de "un", "es" y "binario" de:  

>```ejer_3 = ["Un", "árbol", "binario", "es", "una", "estructura", "de", "un", "tipo", "particular", "a", "veces", "no", "es", "ni", "binario"]```


`````python
ejer_3 = ["Un", "árbol", "binario", "es", "una", "estructura",
          "de", "un", "tipo", "particular", "a", "veces",
          "no", "es", "ni", "binario"]

duplicado = "un"
indices = []
for index, word in enumerate(ejer_3):
    if word.lower() == duplicado.lower():
        indices.append(index)
print(f"Los índices donde aparece '{duplicado}' son: {indices}")

duplicado = "es"
indices = []
for index, word in enumerate(ejer_3):
    if word.lower() == duplicado.lower():
        indices.append(index)
print(f"Los índices donde aparece '{duplicado}' son: {indices}")

duplicado = "binario"
indices = []
for index, word in enumerate(ejer_3):
    if word.lower() == duplicado.lower():
        indices.append(index)
print(f"Los índices donde aparece '{duplicado}' son: {indices}")
`````

## Ejercicio 13
Escribe una función que compruebe si un DNI es válido, devolviendo `True` en caso afirmativo.  
En primer lugar, el DNI debe tener entre 7 y 8 dígitos numéricos.  
Para saber si la letra del DNI es correcta puedes utilizar el siguiente fragmento de código:
>```Python
>DNI= 55555555
>palabra='TRWAGMYFPDXBNJZSQVHLCKE'
>print('La letra del DNI es: ', palabra[DNI%23])
>````


`````python
def dni_valido(dni):
    letras = "TRWAGMYFPDXBNJZSQVHLCKE"
    if len(dni) not in [8, 9] or not dni[:-1].isdigit():
        return False
    return dni[-1].upper() == letras[int(dni[:-1]) % 23]

print(dni_valido("12345678T"))
print(dni_valido("12345678Z"))
`````
